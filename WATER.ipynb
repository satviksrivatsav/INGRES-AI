```python
# ================================
# INGRES-AI ML MODEL TRAINING
# Resource Planner + Safety Checker
# ================================

import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

from google.colab import files

print("Upload the two CSV datasets")
uploaded = files.upload()

# ================================
# LOAD DATASETS
# ================================

resource_df = pd.read_csv("resource_planner_dataset.csv")
safety_df = pd.read_csv("safety_checker_dataset.csv")

print("\nResource Dataset Shape:", resource_df.shape)
print("Safety Dataset Shape:", safety_df.shape)

# ================================
# RESOURCE PLANNER MODEL
# ================================

print("\nTraining Resource Planner Model...")

resource_df = resource_df.copy()

encoders_resource = {}

for col in ["crop", "season", "rainfall", "soil_type"]:
    le = LabelEncoder()
    resource_df[col] = le.fit_transform(resource_df[col])
    encoders_resource[col] = le

target_encoder_resource = LabelEncoder()
resource_df["risk_level"] = target_encoder_resource.fit_transform(resource_df["risk_level"])

X_resource = resource_df.drop(["risk_level", "estimated_water_need"], axis=1)
y_resource = resource_df["risk_level"]

X_train, X_test, y_train, y_test = train_test_split(
    X_resource,
    y_resource,
    test_size=0.2,
    random_state=42
)

resource_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=12,
    random_state=42
)

resource_model.fit(X_train, y_train)

pred = resource_model.predict(X_test)

print("\nResource Planner Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))

joblib.dump(resource_model, "resource_planner_model.pkl")
joblib.dump(encoders_resource, "resource_planner_encoders.pkl")
joblib.dump(target_encoder_resource, "resource_planner_target.pkl")

print("\nResource Planner model saved")

# ================================
# SAFETY CHECKER MODEL
# ================================

print("\nTraining Safety Checker Model...")

safety_df = safety_df.copy()

encoders_safety = {}

for col in ["water_source", "area_type"]:
    le = LabelEncoder()
    safety_df[col] = le.fit_transform(safety_df[col])
    encoders_safety[col] = le

target_encoder_safety = LabelEncoder()
safety_df["safety_status"] = target_encoder_safety.fit_transform(safety_df["safety_status"])

X_safety = safety_df.drop(["safety_status"], axis=1)
y_safety = safety_df["safety_status"]

X_train, X_test, y_train, y_test = train_test_split(
    X_safety,
    y_safety,
    test_size=0.2,
    random_state=42
)

safety_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

safety_model.fit(X_train, y_train)

pred = safety_model.predict(X_test)

print("\nSafety Checker Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))

joblib.dump(safety_model, "safety_checker_model.pkl")
joblib.dump(encoders_safety, "safety_checker_encoders.pkl")
joblib.dump(target_encoder_safety, "safety_checker_target.pkl")

print("\nSafety Checker model saved")

# ================================
# DOWNLOAD MODELS
# ================================

files.download("resource_planner_model.pkl")
files.download("resource_planner_encoders.pkl")
files.download("resource_planner_target.pkl")

files.download("safety_checker_model.pkl")
files.download("safety_checker_encoders.pkl")
files.download("safety_checker_target.pkl")

print("\nAll models downloaded successfully")
```
